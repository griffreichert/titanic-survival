 
### Fill in missing values for age and fare



```python
median_age_by_sex_pclass = train[~pd.isna(train.age)].groupby(['sex', 'pclass']).age.median().rename('median_age')
median_age_by_sex_pclass
```

```python
# fill in NA ages with the median age of people with the same sex and pclass 
#   - Sex & pclass used due to low dimensionality and correlation with age
train['age'] = (
    train['age'].fillna(
        train.merge(median_age_by_sex_pclass, how='left', 
                    left_on=['sex', 'pclass'], right_index=True)['median_age']
    )
)
test['age'] = (
    test['age'].fillna(
        test.merge(median_age_by_sex_pclass, how='left', 
                    left_on=['sex', 'pclass'], right_index=True)['median_age']
    )
)
```

```python
assert sum(pd.isna(test.age)) == 0 and sum(pd.isna(train.age)) == 0
```

```python
median_fare_by_pclass = train[~pd.isna(train.fare)].groupby('pclass').fare.median().rename('median_fare')
median_fare_by_pclass
```

```python
# fill in NA fares with the median fare of people with the same pclass 

train['fare'] = (
    train['fare'].fillna(
        train.merge(median_fare_by_pclass, how='left', 
                    left_on='pclass', right_index=True)['median_fare']
    )
)
test['fare'] = (
    test['fare'].fillna(
        test.merge(median_fare_by_pclass, how='left', 
                    left_on='pclass', right_index=True)['median_fare']
    )
)
```

```python
assert sum(pd.isna(test.fare)) == 0 and sum(pd.isna(train.fare)) == 0
```

## 3. Feature Engineering

```python
# drop the cabin and ticket columns as they are mostly noise
train = train.drop(columns=['cabin', 'ticket'])
test = test.drop(columns=['cabin', 'ticket'])

train.shape, test.shape
```

```python
train.head()
```

### Convert categorical features to numeric

```python
train.groupby('embarked').survived.mean()
```

```python
encoder = TargetEncoder()
train['embarked_target_encoded'] = encoder.fit_transform(train.embarked, train.survived)
test['embarked_target_encoded'] = encoder.transform(test.embarked)
```

```python
embarked_categories = {
    'S': 0,
    'C': 1, 
    'Q': 2, 
}
train['embarked_ordinal_encoded'] = train.embarked.map(embarked_categories)
test['embarked_ordinal_encoded'] = test.embarked.map(embarked_categories)
```

```python
train['sex_target_encoded'] = encoder.fit_transform(train.sex, train.survived)
test['sex_target_encoded'] = encoder.transform(test.sex)
```

```python
gender_categories = {
    'male': 0, 
    'female': 1
}
train['sex_ordinal_encoded'] = train.sex.map(gender_categories)
test['sex_ordinal_encoded'] = test.sex.map(gender_categories)
```

```python
train.head()
```

### Create column 'title' from the name column

```python
get_title = lambda x: re.findall(r"\w+\. ", x)[0].replace('. ', '').lower()
train['title'] = train.name.apply(get_title)
test['title'] = test.name.apply(get_title)
```

```python
title_df = pd.merge(
    train['title'].value_counts(), 
    test['title'].value_counts(), 
    how='outer', 
    left_index=True, 
    right_index=True
).fillna(0).reset_index(level=0)
title_df = title_df.rename(columns={'index':'title'})
title_df['count'] = title_df['title_x'] + title_df['title_y']
title_df['count'] = title_df['count'].astype(int)
title_df = title_df.drop(columns=['title_x', 'title_y'])
title_df
```

```python
title_map = {
    'mlle': 'miss',
    'major': 'misc',
    'col': 'misc',
    'countess': 'misc',
    'capt': 'misc',
    'ms': 'miss',
    'sir': 'mr',
    'lady': 'mrs',
    'mme': 'mrs',
    'don': 'mr',
    'dona': 'mrs',
    'jonkheer': 'misc',
}
train['title'] = train.title.map(title_map).fillna(train.title)
test['title'] = test.title.map(title_map).fillna(test.title)
```

```python
title_df['title'] = title_df.title.map(title_map).fillna(title_df.title)
# title_df.groupby('title')['title'].value_counts()
```

```python
train['title'].value_counts().sort_values().plot(kind='barh')
```

```python
test['title'].value_counts().sort_values().plot(kind='barh')
```

```python
train['title_target_encoded'] = encoder.fit_transform(train.title, train.survived)
test['title_target_encoded'] = encoder.transform(test.title)
```

```python
title_ordinal = {
    'master': 1,
    'mr': 2,
    'rev': 3,
    'dr': 4,
    'mrs': 5,
    'miss': 6,
    'misc': 9,
}
train['title_ordinal_encoded'] = train.title.map(title_ordinal)
test['title_ordinal_encoded'] = test.title.map(title_ordinal)
```

```python
# drop raw categorical features 
train = train.drop(columns=['name', 'sex', 'embarked', 'title'])
test = test.drop(columns=['name', 'sex', 'embarked', 'title'])
train.head()
```
